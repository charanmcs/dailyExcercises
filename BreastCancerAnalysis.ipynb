```python
data['diagnosis'] = data.diagnosis.apply(lambda x :1 if x=='M' else 0)
```

```python
data.drop('id',inplace=True,axis=1)
```

```python
data['area_mean'].corr(data['radius_mean'])
```

```python
data[data.columns].corr()
```

```python
#as we can see perimeter mean,radius mean and area mean have correlation almost equal to 1
data.drop(['area_mean','perimeter_mean'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data_copy=pd.DataFrame(data)
for c in data.columns[1:]:
    data_copy[c] = data[c]  / data[c].abs().max()
```
